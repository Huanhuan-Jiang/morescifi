### Only get books in English

```python
import os
import json
```

```python
cont_dir = '../sfgram/book-contents'
json_dir = '../sfgram/books'
content_file = 'scifi_english.txt'
ids_file = 'enlgish_ids.txt'
```

```python
# read all English books into one txt file and obtain the id for each book
n_books = 1393
count = 0
english_ids = []
with open(ids_file, 'w') as ids, open(content_file, 'w') as books:
    for i in range(1, 1394):
        id = f'{i:05}'
        jsonfn = 'books' + id + '.json'
        json_path = os.path.join(json_dir, jsonfn)

        with open(json_path, 'r') as f:
            data = json.load(f)
        language = data.get("language")
        if language == "English":
            filename = 'book' + id + '.txt'
            file_path = os.path.join(cont_dir, filename)
            if os.path.exists(file_path):
                count += 1
                english_ids.append(id)
                ids.write(id)
                ids.write('\n')

                with open(file_path, 'r') as book:
                    books.write(book.read())
                    books.write('\n') 

print(f'Saved English book ids in {ids_file}')
print(f'There are {count} Enlgish books in the SFGram dataset.')
print(f'Merged files into {content_file}')
```

```python
with open(content_file, 'r', encoding='utf-8') as f:
    text = f.read()
print(f'There are {len(text)} characters in books.')
```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
print(f'There are {vocab_size} different chars in the dataset.')
```

```python
from prettytable import PrettyTable
freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

# Print header
print(f"{'Character':<10} {'Frequency':<10}")
print("-" * 20)

freq_filename = 'char_freq.txt'
with open(freq_filename, 'w') as file:
    for char, count in freq.items():
        print(f"'{char}' {count:<10}")
        #file.write(f"{char} {count}\n")  
```

```python
sorted_freq = dict(sorted(freq.items(), key=lambda item:item[1]))
for char, count in sorted_freq.items():
    print(f"'{char}' {count:<10}")
```

```python
# explore the context of some rare word
def print_context(text, char):
    location = text.find(char)
    print(f'location of {char}: {location}')
    print(text[location-100:location+100])
    print('\n')

i = 0
for char, count in sorted_freq.items():
    if count < 100:
        print_context(text, char)
        i += 1
print(f'There are {i} chars with frequency less than 100.\n')
```

```python
import tiktoken
enc = tiktoken.get_encoding("cl100k_base")
tokens = enc.encode(text)
len_tokens = len(tokens)
print(f'There are {len_tokens} tokens after encoding')

tokens_filename = "tokens.txt"
with open(tokens_filename, 'w') as file:
    file.write(' '.join(map(str, tokens)))
print(f'Tokens have been saved to {tokens_filename}.')
```
